A fun√ß√£o abaixo tem a finalidade de separar dados que comp√µem a camada de entrada tamb√©m conhecido como caracter√≠sticas das informa√ß√µes que representam a resposta. O livro chama esses dois conjuntos de $Œ©^{(x)}$ e $Œ©^{D}$ respectivamente.

```python
def data_split(data):
  omega_d = list()
  omega_x = list()

  for am in data:
    amostra = [-1]
    amostra.extend(am)
    omega_d.append(amostra[-1])
    omega_x.append(amostra[:-1])

  omega_x = np.array(omega_x)
  omega_d = np.array(omega_d)

  return omega_x, omega_d
```

## Perceptron

A implementa√ß√£o do perceptron abaixo, foi baseada nas intru√ßoes do livro: Redes Neurais Artificiais para Engenharia e Ci√™ncias Aplicadas.

A atualiza√ß√£o de pesos pode ser expressada pela equa√ß√£o algoritma:

$W ‚Üê W + ùëõ ‚ãÖ (d^{(k)} - y) ‚ãÖ x^{(k)}$

sendo:

W: o vetor contendo o limiar e os pesos  \\
ùëõ: taxa de aprendizagem da rede \\
$d^{(k)}$: valor desejado de saida para a amostra k \\
y: saida predita pelo perceptron \\
$x^{(k)}$: amostra na posi√ß√£o k \\

```python
class perceptron:
  def __init__(self, n_features, lr = 0.01):
    self.__n_features = n_features
    self.__lr = lr
    self.__epoch = 0
    self.__erro = True
    self.__W = list()
    self.__sys = list()

    for i in range(self.__n_features + 1):
      self.__W.append(random())

  def degree(self, num):
    return (1 if num >= 0 else 0)

  def bipolar_degree(self, num):
    if num > 0:
      num = 1

    elif num < 0:
      num = -1

    return num

  def train(self, X_train, y_train, activation_func = 'bipolar'):
    self.__epoch = 0
    self.__sys.append(np.array(self.W))
    while(self.__erro):
      k = 0

      self.__erro = False
      for amostra in X_train:
        u = sum(self.__W * amostra.T)

        if activation_func == 'bipolar':
          y_hat = self.bipolar_degree(u)

        if activation_func == 'degree':
          y_hat = self.degree(u)

        if y_hat != y_train[k]:
          self.__W = self.__W + (self.__lr*(y_train[k] - y_hat)*amostra)
          self.__sys.append(self.__W)
          self.__erro = True

        k += 1

      self.__epoch+=1

    return self.__W, self.__epoch

  def plot_sys(self):
    columns = ['W'+str(i) for i in range(self.__n_features + 1)]
    sys = pd.DataFrame(self.__sys, columns = columns)

    plt.figure(figsize = (15,7))
    plt.title('Valor dos pesos ao decorrer das modifica√ß√µes.')

    for i in range(self.__n_features + 1):
      plt.plot(np.arange(0,len(self.__sys)), sys[columns[i]], label=columns[i])

    plt.legend(loc = "upper right")
    plt.show()

  def predict(self, X_val, activation_func = 'bipolar'):
    pred = list()
    for amostra in X_val:
      amostra = np.array(amostra)
      u = sum(self.W * amostra.T)

      if activation_func == 'bipolar':
        y_hat = self.bipolar_degree(u)

      if activation_func == 'degree':
        y_hat = self.degree(u)

      pred.append(y_hat)

    return pred


  @property
  def W(self):
    return self.__W
```

```python
Perceptron = perceptron(3)
W_iniciais = Perceptron.W

omega_x, omega_d = data_split(data_perceptron)

W, epocas = Perceptron.train(omega_x, omega_d)

Perceptron.plot_sys()

print(f"< N√∫mero de ep√≥cas: {epocas} >")
print(f"< Pesos iniciais: {W_iniciais} >")
print(f"< Pesos finais: {W} >")
```

### Conclus√£o

A rede perceptron √© um modelo de rede neural que conta com uma grande simplicidade de implementa√ß√£o e entendimento. O modelo √© um classificador bin√°rio, ele separa os dados em duas classes desde de que sejam linearmente separ√°veis, a rede perceptron apenas funciona para classifica√ß√µes onde os dados possam ser separados por uma linha reta ou um hiperplano em dimens√µes mais altas como √© o caso do exerc√≠cio feito neste notebook. Ele n√£o √© capaz de aprender fun√ß√µes n√£o lineares complexas.

Devido ao seu crit√©rio de parada, este modelo n√£o permite que o treinamento pare enquanto o erro n√£o seja nulo. Isto causa que a acur√°cia de treino sempre seja igual a 100%, que √© um indicativo de overfitting, isto pode se agravar dependendo da quantidade de dados, al√©m provocar uma sensibilidade a outliers, uma vez que o algoritmo tem que se ajustar a todos os dados incluindo eles.

Embora ele seja um classificador linear, pode acontecer de ele n√£o convergir para uma resposta que separe os dados perfeitamente, e nunca atingir o crit√©rio de parada. Isso pode acontecer devido a alguns motivos, sendo eles: a rela√ß√£o dos dados ser complexa, a taxa de aprendizagem ser alta ou devido a falta de normaliza√ß√£o dos dados.



## Adaline

A implementa√ß√£o do adaline abaixo, foi baseada nas intru√ßoes do livro: Redes Neurais Artificiais para Engenharia e Ci√™ncias Aplicadas.

A atualiza√ß√£o de pesos pode ser expressada pela equa√ß√£o algoritma:

$W ‚Üê W + ùëõ ‚ãÖ (d^{(k)} - u) ‚ãÖ x^{(k)}$

sendo:

W: o vetor contendo o limiar e os pesos  \\
ùëõ: taxa de aprendizagem da rede \\
$d^{(k)}$: valor desejado de saida para a amostra k \\
u: Pot√™ncial de ativa√ß√£o ou saida do combinador linear \\
$x^{(k)}$: amostra na posi√ß√£o k \\

```python
class adaline(perceptron):
  def __init__(self, n_features, lr = 0.01, epsilon = 0.000001):
    self.epsilon = epsilon
    self.__n_features = n_features
    self.__lr = lr
    self.__epoch = 0
    self.__erro = True
    self.__W = list()
    self.__sys = list()

    for i in range(self.__n_features + 1):
      self.__W.append(random())

  def train(self, X_train, y_train, activation_func = 'bipolar'):
    self.__epoch = 0
    self.__sys.append(np.array(self.__W))
    EQM_anterior = 0
    EQM = 0
    self.p = X_train.shape[0]

    while(True):
      k = 0
      EQM_anterior = EQM
      EQM = 0

      for amostra in X_train:
        u = sum(self.__W * amostra.T)
        EQM += ((y_train[k] - u) ** 2)

        self.__W = self.__W + (self.__lr*(y_train[k] - u)*amostra)
        self.__sys.append(self.__W)
        self.__erro = True

        k += 1

      self.__epoch+=1

      EQM /= self.p

      if abs(EQM - EQM_anterior) <= self.epsilon or (self.__epoch > 1000):
        break

    return self.__W, self.__epoch

  def plot_sys(self):
    columns = ['W'+str(i) for i in range(self.__n_features + 1)]
    sys = pd.DataFrame(self.__sys, columns = columns)

    plt.figure(figsize = (15,7))
    plt.title('Valor dos pesos ao decorrer das modifica√ß√µes.')

    for i in range(self.__n_features + 1):
      plt.plot(np.arange(0,len(self.__sys)), sys[columns[i]], label=columns[i])

    plt.legend(loc = "upper right")
    plt.show()

  def predict(self, X_val, activation_func = 'bipolar'):
    pred = list()
    for amostra in X_val:
      amostra = np.array(amostra)
      u = sum(self.W * amostra.T)

      if activation_func == 'bipolar':
        y_hat = self.bipolar_degree(u)

      if activation_func == 'degree':
        y_hat = self.degree(u)

      pred.append(y_hat)

    return pred

  @property
  def W(self):
    return self.__W
```
